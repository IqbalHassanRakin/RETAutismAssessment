```python
user_files = []

files = glob.glob(url_data_loc + '/' + '*nter*')
for f in files:
    user_files.append(f)

print('No. of files ' + '(User ' + user + ') : ' + str(len(user_files)))
```

```python
f_ip = []
l_ip = []
r_ip = []
t_ip = []
b_ip = []

f_p1 = [400, 0, 0]
f_p2 = [-400, 0, 0]
f_p3 = [-400, -600, 0]
f_p4 = [400, -600, 0]

l_p1 = [-400, 0, 0]
l_p2 = [-400, -600, 0]
l_p3 = [-400, -600, 600]
l_p4 = [-400, 0, 600]

r_p1 = [400, 0, 0]
r_p2 = [400, -600, 0]
r_p3 = [400, -600, 600]
r_p4 = [400, 0, 600]

t_p1 = [400, 0, 0]
t_p2 = [400, 0, 600]
t_p3 = [-400, 0, 600]
t_p4 = [-400, 0, 0]

b_p1 = [400, -600, 0]
b_p2 = [400, -600, 600]
b_p3 = [-400, -600, 600]
b_p4 = [-400, -600, 0]

f_e1 = np.array(f_p2) - np.array(f_p1)
l_e1 = np.array(l_p2) - np.array(l_p1)
r_e1 = np.array(r_p2) - np.array(r_p1)
t_e1 = np.array(t_p2) - np.array(t_p1)
b_e1 = np.array(b_p2) - np.array(b_p1)

f_e2 = np.array(f_p4) - np.array(f_p1)
l_e2 = np.array(l_p4) - np.array(l_p1)
r_e2 = np.array(r_p4) - np.array(r_p1)
t_e2 = np.array(t_p4) - np.array(t_p1)
b_e2 = np.array(b_p4) - np.array(b_p1)

f_normal = np.cross(f_e1, f_e2)
f_point_on_plane = np.array(f_p1)
l_normal = np.cross(l_e1, l_e2)
l_point_on_plane = np.array(l_p1)
r_normal = np.cross(r_e1, r_e2)
r_point_on_plane = np.array(r_p1)
t_normal = np.cross(t_e1, t_e2)
t_point_on_plane = np.array(t_p1)
b_normal = np.cross(b_e1, b_e2)
b_point_on_plane = np.array(b_p1)

for n, f in enumerate(user_files):
  data = dataset.open(f)
  df = data.to_dataFrame()
  print('File : ' + str(n + 1), '|| Number of Data Points : ' + str(df.shape[0]), '|| Task Index: ' + str(df['task_index'].iloc[0]), '|| Task Name : ' + str(df['task_ability'].iloc[0]), '|| Difficulty Level : ' + str(df['task_difficultyLevel'].iloc[0]))

  for i in range(len(df['eye_gaze_rx'])):
    vector = np.array([df['eye_gaze_rx'].iloc[i], df['eye_gaze_ry'].iloc[i], df['eye_gaze_rz'].iloc[i]])
    point_on_vector = np.array([df['skeleton_head_x'].iloc[i], df['skeleton_head_y'].iloc[i], df['skeleton_head_z'].iloc[i]])
    vector_direction = vector / np.linalg.norm(vector)

    if (np.dot(vector_direction, f_normal) < 0):
      t = np.dot((f_point_on_plane - point_on_vector), f_normal) / np.dot(vector_direction, f_normal)
      intersection_point = point_on_vector + t * vector_direction
      if (np.isnan(intersection_point).any()):
        print('SKIP')
      else:
        f_ip.append(intersection_point)
    else:
      print('SKIP')

    if (np.dot(vector_direction, l_normal) < 0):
      t = np.dot((l_point_on_plane - point_on_vector), l_normal) / np.dot(vector_direction, l_normal)
      intersection_point = point_on_vector + t * vector_direction
      if (np.isnan(intersection_point).any()):
        print('SKIP')
      else:
        l_ip.append(intersection_point)
    else:
      print('SKIP')

    if (np.dot(vector_direction, r_normal) < 0):
      t = np.dot((r_point_on_plane - point_on_vector), r_normal) / np.dot(vector_direction, r_normal)
      intersection_point = point_on_vector + t * vector_direction
      if (np.isnan(intersection_point).any()):
        print('SKIP')
      else:
        r_ip.append(intersection_point)
    else:
      print('SKIP')

    if (np.dot(vector_direction, t_normal) < 0):
      t = np.dot((t_point_on_plane - point_on_vector), t_normal) / np.dot(vector_direction, t_normal)
      intersection_point = point_on_vector + t * vector_direction
      if (np.isnan(intersection_point).any()):
        print('SKIP')
      else:
        t_ip.append(intersection_point)
    else:
      print('SKIP')

    if (np.dot(vector_direction, b_normal) < 0):
      t = np.dot((b_point_on_plane - point_on_vector), b_normal) / np.dot(vector_direction, b_normal)
      intersection_point = point_on_vector + t * vector_direction
      if (np.isnan(intersection_point).any()):
        print('SKIP')
      else:
        b_ip.append(intersection_point)
    else:
      print('SKIP')

  f_X = []
  f_Y = []
  l_X = []
  l_Y = []
  r_X = []
  r_Y = []
  t_X = []
  t_Y = []
  b_X = []
  b_Y = []

  for i in range(len(f_ip)):
    if ((-400 <= f_ip[i][0] <= 400) and (-600 <= f_ip[i][1] <= 0)):
      f_X.append(f_ip[i][0])
      f_Y.append(f_ip[i][1])
  f_X = np.array(f_X)
  f_Y = np.array(f_Y)
  f_a = [400, -400, -400, 400, 400]
  f_b = [0, 0, -600, -600, 0]

  for i in range(len(l_ip)):
    if ((0 <= l_ip[i][2] <= 600) and (-600 <= l_ip[i][1] <= 0)):
      l_X.append(l_ip[i][2])
      l_Y.append(l_ip[i][1])
  l_X = np.array(l_X)
  l_Y = np.array(l_Y)
  l_a = [0, 0, 600, 600, 0]
  l_b = [0, -600, -600, 0, 0]

  for i in range(len(r_ip)):
    if ((0 <= r_ip[i][2] <= 600) and (-600 <= r_ip[i][1] <= 0)):
      r_X.append(r_ip[i][2])
      r_Y.append(r_ip[i][1])
  r_X = np.array(r_X)
  r_Y = np.array(r_Y)
  r_a = [0, 0, 600, 600, 0]
  r_b = [0, -600, -600, 0, 0]

  for i in range(len(t_ip)):
    if ((-400 <= t_ip[i][0] <= 400) and (0 <= t_ip[i][2] <= 600)):
      t_X.append(t_ip[i][0])
      t_Y.append(t_ip[i][2])
  t_X = np.array(t_X)
  t_Y = np.array(t_Y)
  t_a = [400, 400, -400, -400, 400]
  t_b = [0, 600, 600, 0, 0]

  for i in range(len(b_ip)):
    if ((-400 <= b_ip[i][0] <= 400) and (0 <= b_ip[i][2] <= 600)):
      b_X.append(b_ip[i][0])
      b_Y.append(b_ip[i][2])
  b_X = np.array(b_X)
  b_Y = np.array(b_Y)
  b_a = [400, 400, -400, -400, 400]
  b_b = [0, 600, 600, 0, 0]

  c = [95, -185, -185, 95, 95]
  d = [-20, -20, -600, -600, -20]

  plt.figure(figsize=(8, 6), dpi=100)
  sns.set_style("whitegrid", {'axes.grid' : False})
  sns.kdeplot(f_X, f_Y, fill=True)
  plt.xlim(-400, 400)
  plt.ylim(-600, 0)
  plt.plot(f_a, f_b, color="red", linewidth=3)
  plt.plot(c, d, color="green", linewidth=3)
  plt.tight_layout()
  plt.xticks([])
  plt.yticks([])
  plt.savefig(url_sm_loc + '/' + str(df['task_index'].iloc[0]) + '_' + str(df['condition'].iloc[0]) + '_f_' + str(df['task_ability'].iloc[0]) + '_' + str(df['task_difficultyLevel'].iloc[0]) + '_' + str(df['ados_preTest_total'].iloc[0]))

  plt.figure(figsize=(8, 6), dpi=100)
  sns.set_style("whitegrid", {'axes.grid' : False})
  sns.kdeplot(l_X, l_Y, fill=True)
  plt.xlim(0, 600)
  plt.ylim(-600, 0)
  plt.plot(l_a, l_b, color="red", linewidth=3)
  plt.tight_layout()
  plt.xticks([])
  plt.yticks([])
  plt.savefig(url_sm_loc + '/' + str(df['task_index'].iloc[0]) + '_' + str(df['condition'].iloc[0]) + '_l_' + str(df['task_ability'].iloc[0]) + '_' + str(df['task_difficultyLevel'].iloc[0]) + '_' + str(df['ados_preTest_total'].iloc[0]))

  plt.figure(figsize=(8, 6), dpi=100)
  sns.set_style("whitegrid", {'axes.grid' : False})
  sns.kdeplot(r_X, r_Y, fill=True)
  plt.xlim(0, 600)
  plt.ylim(-600, 0)
  plt.plot(r_a, r_b, color="red", linewidth=3)
  plt.tight_layout()
  plt.xticks([])
  plt.yticks([])
  plt.savefig(url_sm_loc + '/' + str(df['task_index'].iloc[0]) + '_' + str(df['condition'].iloc[0]) + '_r_' + str(df['task_ability'].iloc[0]) + '_' + str(df['task_difficultyLevel'].iloc[0]) + '_' + str(df['ados_preTest_total'].iloc[0]))

  plt.figure(figsize=(8, 6), dpi=100)
  sns.set_style("whitegrid", {'axes.grid' : False})
  sns.kdeplot(t_X, t_Y, fill=True)
  plt.xlim(-400, 400)
  plt.ylim(0, 600)
  plt.plot(t_a, t_b, color="red", linewidth=3)
  plt.tight_layout()
  plt.xticks([])
  plt.yticks([])
  plt.savefig(url_sm_loc + '/' + str(df['task_index'].iloc[0]) + '_' + str(df['condition'].iloc[0]) + '_t_' + str(df['task_ability'].iloc[0]) + '_' + str(df['task_difficultyLevel'].iloc[0]) + '_' + str(df['ados_preTest_total'].iloc[0]))

  plt.figure(figsize=(8, 6), dpi=100)
  sns.set_style("whitegrid", {'axes.grid' : False})
  sns.kdeplot(b_X, b_Y, fill=True)
  plt.xlim(-400, 400)
  plt.ylim(0, 600)
  plt.plot(b_a, b_b, color="red", linewidth=3)
  plt.tight_layout()
  plt.xticks([])
  plt.yticks([])
  plt.savefig(url_sm_loc + '/' + str(df['task_index'].iloc[0]) + '_' + str(df['condition'].iloc[0]) + '_b_' + str(df['task_ability'].iloc[0]) + '_' + str(df['task_difficultyLevel'].iloc[0]) + '_' + str(df['ados_preTest_total'].iloc[0]))
```
