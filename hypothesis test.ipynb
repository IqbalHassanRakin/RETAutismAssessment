```python
tasks = []
for root, dirs, files in os.walk('/content/drive/MyDrive/DreamDB/User 3'):
    for f in files:
        if not f.endswith('.json'): continue
        data = dataset.open(os.path.join(root,f))
        t = data["task"]
        t['intervention'] = f.split('_')[2]
        t['sourceFile'] = f
        tasks.append(t)
sessions = pd.DataFrame(tasks)
sessions.sort_values(by='index',inplace=True)
sessions.set_index('index')
```

```python
url_03 = '/content/drive/MyDrive/DreamDB/User 3/'
url_37 = '/content/drive/MyDrive/DreamDB/User 37/'

user_03_files = []
user_37_files = []

files_03 = glob.glob(url_03 + '*nter*')
for f in files_03:
    user_03_files.append(f)

files_37 = glob.glob(url_37 + '*nter*')
for f in files_37:
    user_37_files.append(f)

print('No. of interventtion files (User 03) : ' + str(len(user_03_files)))
print('No. of interventtion files (User 37) : ' + str(len(user_37_files)))
```

```python
li_03 = []

for f in user_03_files:
  data = dataset.open(f)
  df = data.to_dataFrame()
  li_03.append(df)

df_03 = pd.concat(li_03, axis = 0, ignore_index = True)
```

```python
li_37 = []

for f in user_37_files:
  data = dataset.open(f)
  df = data.to_dataFrame()
  li_37.append(df)

df_37 = pd.concat(li_37, axis = 0, ignore_index = True)
```

```python
df_03 = df_03.dropna()
df_37 = df_37.dropna()
```

```python
df_03_IM = df_03.copy()
df_03_TT = df_03.copy()
df_03_JA = df_03.copy()
```

```python
df_37_IM = df_37.copy()
df_37_TT = df_37.copy()
df_37_JA = df_37.copy()
```

```python
df_03_IM = df_03_IM.loc[df_03_IM['task_ability'] == 'IM']
df_03_TT = df_03_TT.loc[df_03_TT['task_ability'] == 'TT']
df_03_JA = df_03_JA.loc[df_03_JA['task_ability'] == 'JA']
```

```python
df_37_IM = df_37_IM.loc[df_37_IM['task_ability'] == 'IM']
df_37_TT = df_37_TT.loc[df_37_TT['task_ability'] == 'TT']
df_37_JA = df_37_JA.loc[df_37_JA['task_ability'] == 'JA']
```

```python
cols = [
 'eye_gaze_rx',
 'eye_gaze_ry',
 'eye_gaze_rz',
 'head_gaze_rx',
 'head_gaze_ry',
 'head_gaze_rz',
 'skeleton_elbow_left_confidence',
 'skeleton_elbow_left_x',
 'skeleton_elbow_left_y',
 'skeleton_elbow_left_z',
 'skeleton_elbow_right_confidence',
 'skeleton_elbow_right_x',
 'skeleton_elbow_right_y',
 'skeleton_elbow_right_z',
 'skeleton_hand_left_confidence',
 'skeleton_hand_left_x',
 'skeleton_hand_left_y',
 'skeleton_hand_left_z',
 'skeleton_hand_right_confidence',
 'skeleton_hand_right_x',
 'skeleton_hand_right_y',
 'skeleton_hand_right_z',
 'skeleton_head_confidence',
 'skeleton_head_x',
 'skeleton_head_y',
 'skeleton_head_z',
 'skeleton_sholder_center_confidence',
 'skeleton_sholder_center_x',
 'skeleton_sholder_center_y',
 'skeleton_sholder_center_z',
 'skeleton_sholder_left_confidence',
 'skeleton_sholder_left_x',
 'skeleton_sholder_left_y',
 'skeleton_sholder_left_z',
 'skeleton_sholder_right_confidence',
 'skeleton_sholder_right_x',
 'skeleton_sholder_right_y',
 'skeleton_sholder_right_z',
 'skeleton_wrist_left_confidence',
 'skeleton_wrist_left_x',
 'skeleton_wrist_left_y',
 'skeleton_wrist_left_z',
 'skeleton_wrist_right_confidence',
 'skeleton_wrist_right_x',
 'skeleton_wrist_right_y',
 'skeleton_wrist_right_z']
```

```python
from scipy import stats
```

```python
p_IM = pd.DataFrame(columns = ['T-Test Statistic', 'T-Test P Value', 'ANOVA Statistic', 'ANOVA P Value'])

for c in cols:
  a, b = stats.ttest_ind(df_03_IM[c], df_37_IM[c])
  c, d = stats.f_oneway(df_03_IM[c], df_37_IM[c])
  p_IM = p_IM.append({'T-Test Statistic' : a, 'T-Test P Value' : b, 'ANOVA Statistic' : c, 'ANOVA P Value' : d}, ignore_index = True)

print(p_IM.info())
```

```python
p_TT = pd.DataFrame(columns = ['T-Test Statistic', 'T-Test P Value', 'ANOVA Statistic', 'ANOVA P Value'])

for c in cols:
  a, b = stats.ttest_ind(df_03_TT[c], df_37_TT[c])
  c, d = stats.f_oneway(df_03_TT[c], df_37_TT[c])
  p_TT = p_TT.append({'T-Test Statistic' : a, 'T-Test P Value' : b, 'ANOVA Statistic' : c, 'ANOVA P Value' : d}, ignore_index = True)

print(p_TT.info())
```

```python
p_JA = pd.DataFrame(columns = ['T-Test Statistic', 'T-Test P Value', 'ANOVA Statistic', 'ANOVA P Value'])

for c in cols:
  a, b = stats.ttest_ind(df_03_JA[c], df_37_JA[c])
  c, d = stats.f_oneway(df_03_JA[c], df_37_JA[c])
  p_JA = p_JA.append({'T-Test Statistic' : a, 'T-Test P Value' : b, 'ANOVA Statistic' : c, 'ANOVA P Value' : d}, ignore_index = True)

print(p_JA.info())
```

```python
p_IM = p_IM.dropna()
p_TT = p_TT.dropna()
p_JA = p_JA.dropna()
```
